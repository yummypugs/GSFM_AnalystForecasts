```python
df = pd.read_csv("Dataframes/refinitiv_output.csv")
df["Date"] = pd.to_datetime(df["Date"]).dt.tz_localize(None)
df
```

# Whole Dataframe  
Add industry group names to all rows

```python
df["GICS Industry Group Name"].replace('', pd.NaT, inplace=True)
df['GICS Industry Group Name'].fillna(method='ffill', inplace=True)
#df.value_counts("GICS Industry Group Name")
#df
```

#### Unifying the quarters

```python
df['New Date'] = pd.Series()

for i in range(len(df)):
    if df.loc[i, "Date"].month == 1:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=1, day=1)
    elif df.loc[i, "Date"].month == 2:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=1, day=1)
    elif df.loc[i, "Date"].month == 3:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=1, day=1)
    elif df.loc[i, "Date"].month == 4:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=4, day=1)
    elif df.loc[i, "Date"].month == 5:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=4, day=1)
    elif df.loc[i, "Date"].month == 6:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=4, day=1)
    elif df.loc[i, "Date"].month == 7:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=7, day=1)
    elif df.loc[i, "Date"].month == 8:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=7, day=1)
    elif df.loc[i, "Date"].month == 9:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=7, day=1)
    elif df.loc[i, "Date"].month == 10:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=10, day=1)
    elif df.loc[i, "Date"].month == 11:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=10, day=1)
    elif df.loc[i, "Date"].month == 12:
        df.loc[i, "New Date"] = df.loc[i, "Date"].replace(month=10, day=1)

df.insert(2, "New Date", df.pop("New Date"))
df["New Date"] = pd.to_datetime(df["New Date"]).dt.date
#df
```

```python
df = df.drop("Date", axis = 1)
df.rename(columns={"New Date": "Date"}, inplace=True)
df["Date"] = pd.to_datetime(df["Date"])
df = df.iloc[::-1].reset_index(drop=True)
#df
```

```python
df = df.dropna(subset="Earnings Per Share - Actual Surprise")
df
```

#### Adding absolute surprise values column (dependent variable)

```python
df["Earnings Per Share - Actual Surprise AbsVals"] = df["Earnings Per Share - Actual Surprise"].abs()
df.insert(5, "Earnings Per Share - Actual Surprise AbsVals", df.pop("Earnings Per Share - Actual Surprise AbsVals"))
#df
```

### Loss Firm Status

```python
df["Net Income after Tax"] = df["Net Income after Tax"].fillna(1)
df["loss firm status"] = [-1 if x > 1 else 1 if x <= 0 else 0 for x in df["Net Income after Tax"]]
#df.value_counts("loss firm status")
#df
```

### Analyst Change

```python
df["Recommendation change"] = df["Recommendation - Mean (1-5)"] - df["Recommendation - Mean (1-5).1"]
df.insert(13, "Recommendation change", df.pop("Recommendation change"))
#df
```

#### Saving as a CSV WITHOUT outliers

```python
# Calculate the 0.15th and 99.85th percentile
pct_0_15 = np.percentile(df['Earnings Per Share - Actual Surprise'], 0.5)
pct_99_85 = np.percentile(df['Earnings Per Share - Actual Surprise'], 99.5)

# Remove the data points outside the 0.15th and 99.85th percentile
df = df[(df['Earnings Per Share - Actual Surprise'] >= pct_0_15) & (df['Earnings Per Share - Actual Surprise'] <= pct_99_85)].copy()
#df
```
